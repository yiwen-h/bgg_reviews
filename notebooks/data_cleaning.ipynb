```python
import pandas as pd
```

```python
dataset = pd.read_csv('../raw_data/bgg-19m-reviews.csv')
```

```python
pd.options.display.max_columns = None
```

```python
dataset.shape
```

```python
dataset.head()
```

```python
dataset = dataset.dropna()
```

```python
dataset.shape
```

```python
game_info = pd.read_csv('../raw_data/games_detailed_info.csv', low_memory = False)
```

```python
game_info2 = game_info[['id', 'yearpublished', 'maxplayers', 'playingtime']]
```

```python
game_info2 = game_info2.rename(columns = {'id': 'ID'})
game_info2.head()
```

```python
dataset.merge(game_info2, how='left', on = 'ID')
```

```python
dataset.to_csv('../raw_data/clean_data.csv')
```

```python
dataset['rating']=dataset['rating'].round(decimals=0)
```

```python
low_ratings=dataset[dataset['rating'] < 5].sample(n=50000, random_state=1)
```

```python
high_ratings=dataset[dataset['rating'] >= 5].sample(n=50000, random_state=1)
```

```python
combined = pd.concat([low_ratings, high_ratings], axis = 0)
combined.shape
```

```python
combined.to_csv('../raw_data/clean_data.csv')
```
